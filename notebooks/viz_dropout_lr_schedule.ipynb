```python
import pandas as pd
```

```python
args_stepFrac = 0.5  # Training iteration when dropout = pDropMax
args_pDropMax = 0.2  # Maximum dropout probability
args_numEpochs = 2
args_numArchives = 84
totalSteps = args_numEpochs * args_numArchives

dropout = []
step = 0
for step in range(totalSteps):

    if 1.0*step < args_stepFrac*totalSteps:  # growing p_drop
        p_drop = args_pDropMax*step/(args_stepFrac*totalSteps)
    else:  # shrinking p_drop
        p_drop = max(0,args_pDropMax*(2*step - totalSteps*(args_stepFrac+1))/(totalSteps*(args_stepFrac-1))) # fast decay
    dropout.append(p_drop)

pd.DataFrame(dropout)[0].plot()
```
